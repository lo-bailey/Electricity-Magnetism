```python
## Used ChatGBT to write loop, it also suggested using lists for storing results 

import numpy as np
import matplotlib.pyplot as plt

# Constants
q = 1.602E-19  # Charge of proton (Coulombs)
m = 1.673E-27  # Mass of proton (kg)
B = np.array([0, 0, -1E-6])  # Magnetic field in the z-direction (Tesla)
dt = 0.001  # Time step (seconds)
n = 25 # Number of steps

# Initial conditions
x = np.array([0.0, 0.0, 0.0])  # Initial position (m)
v = np.array([1.0, 0.0, 0.0])  # Initial velocity (m/s)

# Lists to store results
x_vals = [x.copy()]
v_vals = [v.copy()]
a_vals = []  # Start empty
F_vals = []  # Start empty

# Iterative loop
for step in range(n):
    # Calculate force 
    F = q * np.cross(v, B)  # Lorentz force
    F_vals.append(F.copy())  # Store force

    # Calculate acceleration
    a = F / m
    a_vals.append(a.copy())  # Store acceleration
   
    # Update position
    x = x + v * dt
    x_vals.append(x.copy())  # Store position
    
    # Update velocity
    v = v + a * dt
    v_vals.append(v.copy())  # Store velocity

# Convert results to NumPy arrays for easier analysis
x_vals = np.array(x_vals)
v_vals = np.array(v_vals)
a_vals = np.array(a_vals)
F_vals = np.array(F_vals)

# Print results of first five steps 
print("Positions (x0, x1, x2, x3, x4):")
print(x_vals[0], "\n", x_vals[1], "\n", x_vals[2], "\n", x_vals[3], "\n", x_vals[4])
print("\nVelocities (v0, v1, v2, v3, v4):")
print(v_vals[0], "\n", v_vals[1], "\n", v_vals[2], "\n", v_vals[3], "\n", v_vals[4])
print("\nAccelerations (a0, a1, a2, a3, a4 ):")
print(a_vals[0], "\n", a_vals[1], "\n", a_vals[2], "\n", a_vals[3], "\n", a_vals[4])
print("\nForces (f0, f1, f2, f3, f4):")
print(F_vals[0], "\n", F_vals[1], "\n", F_vals[2], "\n", F_vals[3], "\n", F_vals[4])

# Plot the trajectory
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_vals[:, 0], x_vals[:, 1], x_vals[:, 2], label="Particle Trajectory")

#Configure plot
ax.set_title("Trajectory of a Proton in a Magnetic Field", fontsize=14)
ax.set_xlabel("x (m)", fontsize=12)
ax.set_ylabel("y (m)", fontsize=12)
ax.set_zlabel("z (m)", fontsize=12)
ax.legend()
plt.show()
```
