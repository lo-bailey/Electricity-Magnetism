## E&M: Standard E1

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
q1 = -1.602E-19 # charge of electron in coulombs
q2 = 1.602E-19 # charge of proton in coulombs
Q = 3.215E-7 # uniform charge of rod in coulombs
k = 8.9E7 # coulombs' constant 
L = 1.5 # length of rod in m

# define position vectors
r1 = np.array([0.2, 1.3, -0.9])
r1_mag = np.linalg.norm(r1)
r1_unit = r1 / r1_mag
print(f"r1 mag: {r1_mag}")
print(f"\nr1 unit: {r1_unit}")

r2 = np.array([-0.3, -0.5, 0.9])
r2_mag = np.linalg.norm(r2)
r2_unit = r2 / r2_mag
print(f"\nr2 mag: {r2_mag}")
print(f"\nr2 unit: {r2_unit}")

# calculate E fields at points 1 & 2
E1_calc = ((k * Q) / r1_mag**2) * r1_unit 
E1 = np.array(E1_calc)

E2_calc = ((k * Q) / r2_mag**2) * r2_unit 
E2 = np.array(E2_calc)

# calculate F at points 1 & 2
F1_calc = q1 * E1
F1 = np.array(F1_calc)

F2_calc = q2 * E2
F2 = np.array(F2_calc)

# print calculated values
print( '\nEletric Field at Point 1:', E1 )
print( '\nForce due to E1 at Point 1:', F1 )
print( '\nEletric Field at Point 2:', E2 )
print( '\nForce due to E2 at Point 2:', F2 )

# create plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# plot uniform charged rod 
rod_start = np.array([-L/2,0, 0])
rod_end = np.array([L/2, 0, 0 ])
ax.plot([rod_start[0], rod_end[0]], [rod_start[1], rod_end[1]], [rod_start[2], rod_end[2]], c='red', lw=5, label="Charged Rod")

# plot point charges
ax.scatter(xs=r1[0], ys=r1[1], zs=r1[2], s=100, c='blue', depthshade=True)
ax.scatter(xs=r2[0], ys=r2[1], zs=r2[2], s=100, c='red', depthshade=True)

# plot force vector 
ax.quiver(r1[0], r1[1], r1[2], F1[0], F1[1], F1[2], color='green', length=0.75, normalize=True, label="Force Vector")
ax.quiver(r2[0], r2[1], r2[2], F2[0], F2[1], F2[2], color='green', length=0.75, normalize=True)
ax.quiver(r1[0], r1[1], r1[2], E1[0], E1[1], E1[2], color='purple', length=1.0, normalize=True, label="E-Field Vector")
ax.quiver(r2[0], r2[1], r2[2], E2[0], E2[1], E2[2], color='purple', length=1.0, normalize=True)

#Configure plot
ax.set_title("\nForce on Two Charged Particles Due to Uniformly Charged Rod", fontsize=14)
ax.set_xlabel("x (m)", fontsize=12)
ax.set_ylabel("y (m)", fontsize=12)
ax.set_zlabel("z (m)", fontsize=12)
ax.legend()
plt.show()
```
